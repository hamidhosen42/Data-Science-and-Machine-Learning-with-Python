```python
import pandas as pd
```

```python
df=pd.read_csv("One-Hot Encoding in Python with Pandas Data Set/carprices.csv")
df
```

```python
dummies=pd.get_dummies(df["Car Model"])
dummies
```

```python
dummies_variable=pd.get_dummies(df["Car Model"],drop_first=True)
dummies_variable
```

```python
marged=pd.concat([df,dummies],axis=1)
marged
```

```python
new_df=marged.drop('Car Model',axis=1)
new_df
```

```python
y=df["Sell Price"]
y
```

```python
x=new_df.drop("Sell Price",axis=1)
x
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.score(x,y)
```

```python
model.predict([[69000,6,0,1,0,0]]) #18000
```
