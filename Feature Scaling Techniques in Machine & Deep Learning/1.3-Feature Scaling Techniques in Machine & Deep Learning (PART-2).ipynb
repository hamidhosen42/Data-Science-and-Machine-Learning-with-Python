```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_set=pd.read_csv("1.1-Feature Scaling Techniques in Machine & Deep Learning (PART-1)/feature scaling.csv")
```

```python
data_set
```

```python
x=data_set.iloc[:,1:3]
x
```

#### MIN MAX SCALER or Normalization (Data Preprocessing)

```python
from sklearn.preprocessing import MinMaxScaler 
```

```python
object=MinMaxScaler(feature_range=(0,1))
```

```python
x_after_min_max_scaler=object.fit_transform(x)
```

```python
x_after_min_max_scaler
```

### standardisation 

```python
from sklearn.preprocessing import StandardScaler
```

```python
object=StandardScaler()
```

```python
x_after_Standardisation=object.fit_transform(x)
```
