## Feature Selection Techniques -> Univariate Selection
Score Function use

```python
from sklearn.feature_selection import SelectKBest
```

```python
from sklearn.feature_selection import f_classif
```

```python
object=SelectKBest(score_func=f_classif)
```

```python
object.fit(x,y)# x=feature ,y=label
```

```python
pd.DataFrame(object.scores_)
```

```python
SCORE_COL=pd.DataFrame(object.scores_,columns=['score'])
```

```python
SCORE_COL
```

```python
NAME_COL=pd.DataFrame(x.columns)
NAME_COL
```

```python
top_feature=pd.concat([NAME_COL,SCORE_COL],axis=1)
top_feature
```

```python
top_feature.nlargest(8,'score')
```
