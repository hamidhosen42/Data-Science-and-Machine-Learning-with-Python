# Stemming

```python
# pip install nltk
from nltk.stem import PorterStemmer
```

```python
port = PorterStemmer()
```

```python
words = ['change', 'changing','changed','changes']
```

```python
words
```

```python
for word in words:
    print('Before =', word)
    print('After =',port.stem(word))
    print('\n')
```

# Lemmatization

```python
from nltk.stem import WordNetLemmatizer
```

```python
lem = WordNetLemmatizer()
```

```python
for word in words:
    print('Before =', word)
    print('After =',lem.lemmatize(word))
```
