### First Step

```python
m=0
c=0
l=0.0001
```

```python
n=len(x)
itr=60

for i in range(itr):
    predy=m*x+c
    Dm=(-2/n)*sum(x*(y-predy))
    Dc=(-2/n)*sum(y-predy)
    
    #update
    m=m-l*Dm
    c=c-l*Dc
    
    print(m,c)
```

```python
pred2=m*2000+c
```
